```python
import pandas as pd

json = pd.read_json ('https://pomber.github.io/covid19/timeseries.json')
json
```

```python
for col in json.columns: 
    print(col)
```

```python
country = 'US'
json[country]
```

```python
df = pd.json_normalize(json[country])
df
```

```python
df.dtypes
```

```python
df[["date"]] = df[["date"]].apply(pd.to_datetime)
```

```python
df.dtypes
```

```python
df.date
```

```python
import matplotlib.pyplot as plt

plt.plot(df.confirmed)
```

```python
diff = df.diff()[1:]
df = df[1:]
print(diff)
```

```python
plt.plot(diff.confirmed)
plt.plot(diff.recovered,c='g')
plt.plot(diff.deaths,c='r')
plt.show()
```

```python
plt.scatter(df.confirmed,diff.confirmed)
```

```python
import numpy as np
model = np.poly1d(np.polyfit(df.confirmed, diff.confirmed, 3))
print(model)
```

```python
plt.scatter(df.confirmed,diff.confirmed)
xp = np.linspace(df.confirmed.min(), df.confirmed.max(), 100)
plt.plot(xp, model(xp), c='r')
plt.show()
```
